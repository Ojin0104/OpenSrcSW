```python
import csv
f=open('seoul.csv')
data=csv.reader(f)
header=next(data)
max_tempdif=-999
max_temp=-999
min_temp=999
max_datedif=''
max_date=''
min_date=''
for row in data:
    if row[-1]==''or row[-2]=='':
        row[-1]=-999
        row[-2]=-999
        
    row[-1]=float(row[-1])
    row[-2]=float(row[-2])
    if max_tempdif<(row[4]-row[3]):
        max_datedif=row[0]
        max_tempdif=row[4]-row[3]
    if max_temp<row[4]:
        max_date=row[0]
        max_temp=row[4]
    if row[-1]==-999 or row[-2]==-999:
        row[-1]=100
        row[-2]=100
    if min_temp>row[3]:
        min_date=row[0]
        min_temp=row[3]
f.close()
print('근 한달 서울의 일교차가 가장 큰 날은',max_datedif+'로, 일교차는',max_tempdif,'도 였습니다.')
print('근 한달 서울의 최저기온이 가장 낮은 날은',min_date+'로,',min_temp,'도 였습니다.')
print('근 한달 서울의 최고기온이 가장 높은 날은',max_date+'로,',max_temp,'도 였습니다.')
```
