```python
#-*-coding:utf-8-*-
import csv
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus']=False ##y -   
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[]
result2=[]
for row in data:
    if row[-1]!='':
        if row[0].split('-')[1]=='01'and row[0].split('-')[2]=='04':
            result.append(float(row[-1]))
            result2.append(float(row[-2]))
plt.rc('font',family='Malgun Gothic')##  
plt.plot(result,label=' ')
plt.plot(result2,label=' ')
plt.title(' (1 4) ')
plt.ylabel('(C)')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)
aug=[]
jan=[]
for row in data:
    month=row[0].split('-')[1]
    if row[-1]!='':
        if month=='08' :
            aug.append(float(row[-1]))
        if month=='01':
            jan.append(float(row[-1]))
plt.rc('font',family='Malgun Gothic')##              
plt.hist(aug,bins=100,color='r',label='august')
plt.hist(jan,bins=100,color='b',label='january')
plt.title('1 8  ')
plt.xlabel('(C)')
plt.legend()
plt.show()
```
