```python
#-*-coding:utf-8-*-
import csv
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus']=False ##y축에 -기호 보일 수 있게
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[]
result2=[]
for row in data:
    if row[-1]!='':
        if row[0].split('-')[1]=='01'and row[0].split('-')[2]=='04':
            result.append(float(row[-1]))
            result2.append(float(row[-2]))
plt.rc('font',family='Malgun Gothic')##한글보이게 및 폰트설정
plt.plot(result,label='최고 기온')
plt.plot(result2,label='최저 기온')
plt.title('내 생일(1월 4일)기온 분표')
plt.ylabel('기온(ºC)')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)
aug=[]
jan=[]
for row in data:
    month=row[0].split('-')[1]
    if row[-1]!='':
        if month=='08' :
            aug.append(float(row[-1]))
        if month=='01':
            jan.append(float(row[-1]))
plt.rc('font',family='Malgun Gothic')##한글보이게 및 폰트설정            
plt.hist(aug,bins=100,color='r',label='august')
plt.hist(jan,bins=100,color='b',label='january')
plt.title('1월과 8월 최고기온 분포')
plt.xlabel('최고기온(ºC)')
plt.legend()
plt.show()
```
