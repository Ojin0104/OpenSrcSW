```python
#-*-coding:utf-8-*-
import matplotlib.pyplot as plt
import csv
import pandas as pd
import numpy as np
#encoding='utf-8-sig'
data=pd.read_csv('people.csv',engine='python',encoding='CP949')#한글파일꺠짐해결
rdata=data.T[3:][5]#광진구데이터.T사용해서 행렬 바꾸어 열로 데이터 분석함
gu_num=data.index.stop-1#구 갯수
result=[]
c_result=[]
big=[]#가장많이나는
small=[]#가장적게나는 
for row in rdata:
    result.append(int(row.replace(',',"")))##광진구 데이터 차례로 넣어줌
    
bigsum=0#최대
smallsum=1000000000#최소

for i in range(gu_num):
    if i!=4:
        
        name=data.T[i+1][0]#서울특별시데이터제외
        compare_data=data.T[3:][i+1]#서울특별시 전체데이터 제외하기위해 +1
        t=0#광진구 result리스트 인덱스
        sum=0
        c_result=[]
        for row in compare_data:
            row=int(row.replace(',',""))
            sum+=(result[t]-row)**2
            c_result.append(row)
            t+=1
        if bigsum<sum:
            big=[]
            bigsum=sum
            big=c_result
            bigname=name
        if smallsum>sum:
            small=[]
            smallsum=sum
            small=c_result
            smallname=name
           
        
#print(small)            
        
        
bigS=round(np.std(big),4)#표준편차들
smallS=round(np.std(small),4)#4째자리까지반올림
gwang=round(np.std(result),4)##광진구 표준편차
# f=open('people.csv')
# data=csv.reader(f)
# next(data)
# result=[]
# for row in data:
#     
        
plt.rc('font',family='Malgun Gothic')
plt.style.use('ggplot')
plt.plot(result[2:],label=data.T[5][0][5:10]+str(gwang))
plt.plot(big,label=bigname[5:10]+str(bigS))
plt.plot(small,label=smallname[5:10]+str(smallS))
plt.legend()
plt.title('나이대별 인구현황')
plt.ylabel('인구(명)')
plt.xlabel('나이(세)')
plt.show()
```
