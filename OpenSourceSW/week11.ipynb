```python
class calc:
    def sum(self,a,b):
        result=a+b
        print("%s+%s=%s입니다."%(a,b,result))
    def sub(self,a,b):
        result=a-b
        print("%s-%s=%s입니다."%(a,b,result))
    def multi(self,a,b):
        result=a*b
        print("%s*%s=%s입니다."%(a,b,result))
    def div(self,a,b):
        result=int(a/b)
        print("%s/%s=%s입니다."%(a,b,result))
qq=calc()
qq.sum(3,8)
qq.div(6,3)
qq.sub(2,1)
qq.multi(3,3)
```

```python
class stack:
    def __init__(self):
        self.top=[]
    def __len__(self):#스택 길이반환
        return str(self.top[::1])
    def __str__(self):#스택내부 데이터 string형태로
        return str(self.top[::1])
    def isEmpty(self):
        return len(self.top)==0
    def size(self):
        return len(self.top)
    def push(self,item):
        self.top.append(item)
    def pop(self):
        if not self.isEmpty():
            return self.top.pop()#리스트 마지막 데이터 리턴
        else:
            print("Stack underflow")
            exit()
    def clear(self):
        self.top=[]

NUM=stack()
OP=stack()
cal=input("계산식 입력:")
a=cal.split(" ")
answer=0
for x in a:
    if(x.isdigit()):##숫자면 NUM에 연산자면 OP에 넣어 주고 들어오는 연산자가 +나 -이면 그전에 연산자 값 계산해 준후 push *나 /이면 연산자 push 
        
        NUM.push(int(x))
        
    elif(x=='*' or x=='/'):
        OP.push(x)
    elif(x=='+' or x=='-'):
        if(OP.isEmpty()):
            OP.push(x)
        else:   
            while(not OP.isEmpty()):#+나 -가 나오면 앞에 나온 연산자 뒤에서 부터 순서대로 처리해주면 *랑 /가 우선적으로 계산되어진다.
                if(OP.top[OP.size()-1]=='+'):#+나 -면 덧셈뺼셈해줌
                    
                        answer=NUM.pop()+NUM.pop()               
                        OP.pop()            
                        NUM.push(answer)
                elif(OP.top[OP.size()-1]=='-'):
                        answer=NUM.pop()-NUM.pop()
                        OP.pop()        
                        NUM.push(answer)
                elif(OP.top[OP.size()-1]=='*'):
                        answer=NUM.pop()*NUM.pop()
                        OP.pop()
                        NUM.push(answer)
                elif(OP.top[OP.size()-1]=='/'):
                        answer=1/NUM.pop()*NUM.pop()
                        OP.pop()
                        NUM.push(answer)
            OP.push(x)
##마지막 숫자까지 push 되고 for문 끝이 나므로 남아있는 연산자 까지 계산해 준다.

if(OP.top[OP.size()-1]=='+'):#마지막남은 연산자까지 계산
               
                answer=NUM.pop()+NUM.pop()
                OP.pop()               
                NUM.push(answer)
elif(OP.top[OP.size()-1]=='-'):
                answer=-NUM.pop()+NUM.pop()
                OP.pop()
                NUM.push(answer)
elif(OP.top[OP.size()-1]=='*'):
                answer=NUM.pop()*NUM.pop()
                OP.pop()
                NUM.push(answer)
elif(OP.top[OP.size()-1]=='/'):
                answer=1/NUM.pop()*NUM.pop()
                OP.pop()
                NUM.push(answer)
    
print("계산식 출력:",cal,"=",NUM.pop())
```
